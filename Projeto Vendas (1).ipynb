```python
sales_train_all_df = pd.merge(sales_train_df, store_info_df, how = 'inner', on = 'Store')   #Criamos uma variavel e vamos unir uma tabela na outra por meio do ID da loja que se encontra na COLUNA store
```

```python
sales_train_all_df.shape # Vemos aqui que as colunas aumentaram para 17
```

```python
sales_train_all_df.tail()
```

```python
correlations = sales_train_all_df.corr()
f, ax = plt.subplots(figsize = (20,20))
sns.heatmap(correlations, annot=True)
```

```python
#Agora vamos gerar alguns graficos adicionais
```

```python
sales_train_all_df['Year'] = pd.DatetimeIndex(sales_train_all_df['Date']).year # quando não temos um atributo e colocamos um nome o pandas entende que é um novo atributo
```

```python
sales_train_all_df.head()
```

```python
sales_train_all_df['Month'] = pd.DatetimeIndex(sales_train_all_df['Date']).month
sales_train_all_df['Day'] = pd.DatetimeIndex(sales_train_all_df['Date']).day
```

```python
sales_train_all_df.head()
```

```python
axis = sales_train_all_df.groupby('Month')[['Sales']].mean().plot(figsize = (10,5), marker = 'o', color = 'g')
axis.set_title('Média de Vendas por Mês')
```

```python
axis = sales_train_all_df.groupby('Month')[['Customers']].mean().plot(figsize = (10,5), marker = '^', color = 'g')
axis.set_title('Média de Vendas por Mês')
```

```python
axis = sales_train_all_df.groupby('Day')[['Sales']].mean().plot(figsize = (10,5), marker = 'o', color = 'g')
axis.set_title('Média de Vendas por Dia')
```

```python
axis = sales_train_all_df.groupby('Day')[['Sales']].mean().plot(figsize = (10,5), marker = 'o', color = 'g')
axis.set_title('Média de Clientes por Dia')
```

```python
axis = sales_train_all_df.groupby('DayOfWeek')[['Sales']].mean().plot(figsize = (10,5), marker = 'o', color = 'g')
axis.set_title('Média de Clientes por Dia')
```

```python
fig, ax = plt.subplots(figsize = (20,10))
sales_train_all_df.groupby(['Date', 'StoreType']).mean()['Sales'].unstack().plot(ax = ax)
```
